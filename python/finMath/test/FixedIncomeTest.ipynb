Effective duration is measured by the numerical derivative.
$$ D^* = -\left( \frac{1}{P} \right) \frac{dP}{dy} $$

it is estimated as
$$ D^E = \frac{ \left[ P_{-} - P_{+} \right]}{ \left( 2P_{0}\Delta y \right)}  = \frac{ P\left(y_{0} - \Delta y \right) - P\left(y_{0} + \Delta y \right])}{ \left( 2P_{o}\Delta y \right)}$$

Effective convexity is measured by the numerical derivative.
$$ C = \left( \frac{1}{P} \right) \frac{d^2P}{dy^2} $$

it is estimated as
$$ C^E = \frac{ \left[ D_{-} - D_{+} \right]}{ \Delta y}  = \frac {\left [\frac{ P\left(y_{0} - \Delta y \right) - P_{0} }{ \left( P_{o}\Delta y \right)} - \frac{  P_{0}  - P\left(y_{0} + \Delta y \right)}{ \left( P_{o}\Delta y \right)}\right]}{\Delta y}$$

Consider a 30 year zero coupon bond with yield of 6% with a 1% change either up or down

```python
from functions.assetClasses.bond import BondPricing

par_value: float = 100.00
years_to_maturity: int = 30
interest_rate: float = 0.06
coupon_rate: float = 0.00

bond_price: float = BondPricing.bond_pricing_vanilla(par_value,years_to_maturity,coupon_rate,interest_rate)

zeros_duration =  lambda t, y, f, n : t / (1 + y/f)**1
six_percent_duration: float = zeros_duration(years_to_maturity, interest_rate, 2, 1)
print("The 6% Bond Price. ", bond_price, " Duration ", six_percent_duration)

interest_rate = 0.07
bond_price7: float = BondPricing.bond_pricing_vanilla(par_value,years_to_maturity,coupon_rate,interest_rate)

seven_percent_duration: float = zeros_duration(years_to_maturity, interest_rate, 2, 1)
print("The 7% Bond Price. ", bond_price7, " Duration ", seven_percent_duration)

interest_rate = 0.05
bond_price5: float = BondPricing.bond_pricing_vanilla(par_value,years_to_maturity,coupon_rate,interest_rate)

five_percent_duration: float = zeros_duration(years_to_maturity, interest_rate, 2, 1)
print("The 5% Bond Price. ", bond_price5, " Duration ", five_percent_duration)

#The Effective Duration
effective_duration: float = (bond_price5 - bond_price7)/(2*0.01*bond_price)
print(effective_duration)

#The Effective Convexity
effective_convexity : float = ((bond_price5 - bond_price)/(bond_price* 0.01) - (bond_price - bond_price7)/(bond_price*0.01))/0.01
print(effective_convexity)
```
